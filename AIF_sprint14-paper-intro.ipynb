3.2　Sharing Features for PRN and Fast R-CNNに記載されている内容となります。  
3つの手法((i)Alternating traning (ⅱ)Approximate joint training (ⅲ)Non-approximate joint training)が記載されておりますが、最後に結論（まとめ）として4-Step Alternating Trainingが記載されております。この内容により高速化したのだと思います。文中には高速化という単語は出て来ていませんが。  
つまり、畳み込み層を共有して、2つのネットワークを統合したため高速化されたのだと思います。    
    
ステップ1：まず初めに、RPNを訓練します。  
ステップ2：ステップ1RPNで生成された出力を使用して、Fast R-CNNの検出ネットワークを訓練します。  
ステップ3：RPN訓練で初期化した検出ネットワークを使います。しかし、共有畳み込み層を修正し、RPN固有の層だけを微調整します。  
ステップ4：共有畳み込み層を修正しながら、Fast R-CNN固有の層を微調整します。


**(3) One-Stageの手法とTwo-Stageの手法はどう違うのか。**  
One-Stageはクラス固有の検出パイプラインです。  
P10「class-specific detection pipeline,・・・」のセンテンスが根拠です。  

Two-Stageはクラスに依存しないプロトコルとクラス固有の検出で構成される2段階カスケードです。  
P10「two-stage cascade consisting of class-agnostic pro- posals and class-specific detections.」のセンテンスが根拠です。

**(4) RPNとは何か。**  
Region Proposal Network (RPN)とは、検出ネットワークを伴ったフルイメージ畳み込み機能です。
アブストラクト「shares full-image convolutional features with the detection network, thus enabling nearly cost-free region proposals. 」のセンテンスが根拠です。

**(5) RoIプーリングとは何か。**  
バウンディングボックスを予測するものです。
P6「the predicted bounding boxes」センテンスが根拠です。

**(6) Anchorのサイズはどうするのが適切か。**

P6「we use 3 scales with box areas of 1282, 2562, and 5122 pixels, and 3 aspect ratios of 1:1, 1:2, and 2:1.」センテンスが根拠です。
